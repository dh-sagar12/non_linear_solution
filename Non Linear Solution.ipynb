```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize



# Read CSV data into a pandas DataFrame
data = pd.read_csv('data.csv')

data['Index No'] = data['Index No'].replace(to_replace='NR',value=np.nan).astype(float)
data['Index No'].fillna(data['Index No'].mean(), inplace=True)


num_rows = 50


selected_data = data.sample(n=num_rows, random_state=42)


# Extract data from the DataFrame
duration = selected_data['Duration']
index_no = selected_data['Index No']
ytm = selected_data['YTM (%)']

```

```python

# target 
investment_horizon_target, absorb_loss_constraint_target, target_yeild_target  =  8, 1, 3.5


# 4

min_investment_horizon_target =  investment_horizon_target - (investment_horizon_target *0.1)
max_investment_horizon_target = investment_horizon_target +  (investment_horizon_target *0.1)

# absorb_loss_constraint_target =  1


# target_yeild_target =  4.35
min_target_yeild_target =  target_yeild_target - (target_yeild_target *0.1)
max_target_yeild_target = target_yeild_target + (target_yeild_target *0.1)

def get_investment_horizon_score(investment_horizon_result):
    if investment_horizon_result  ==  0:
        return  0 
    elif investment_horizon_result <  5:
        return  1
    
    elif investment_horizon_result < 10:
        return 2
    
    elif investment_horizon_result < 20:
        return 3
    elif investment_horizon_result < 30:
        return 4
    
    else:
        return 5


def get_target_yeild_score(target_yeild_result):
    if target_yeild_result <= 2.04:
        return 1
    elif target_yeild_result <= 3.34:
        return 2
    elif target_yeild_result < 3.99:
        return 3

    elif target_yeild_result <= 4.4:
        return 4
    else:
        return 5
    
def get_absorb_loss_score(absorb_loss_result):
    if absorb_loss_result  <= 5:
        return 5
    elif absorb_loss_result <= 8 :
        return 4
    elif absorb_loss_result <=11:
        return 3

    elif absorb_loss_result <=14:
        return 2

    else:
        return 1

# Objective function to minimize (negative implied risk tolerance)
def objective_function(weights, *args):
    duration, index_no, ytm, b = args
    investment_horizon =  np.sum(duration * (weights))
    ability_to_absorb_loss  =  np.sum(index_no * (weights))
    target_yeild  =  np.sum(ytm * weights)
    
    investment_horizon_score  =  get_investment_horizon_score(investment_horizon)
    absorb_loss_score =  get_absorb_loss_score(ability_to_absorb_loss)
    target_yeild_score =  get_target_yeild_score(target_yeild)
    
    print(investment_horizon_score, absorb_loss_score, target_yeild_score )
        
    implied_risk_tolerance  =  (investment_horizon_score * 0.333) + (absorb_loss_score * 0.333) + (target_yeild_score * 0.333)
    return -implied_risk_tolerance

# Constraints
def constraint1(weights):
    # Investment Horizon constraint (3.6 <= sum(weights * duration) <= 4.4)
    return  max_investment_horizon_target - np.sum(weights * duration) 


def constraint2(weights):
    # Ability to Absorb Losses constraint (sum(weights * index_no) == 1)
    ability_to_absorb_loss  =  np.sum((weights) * index_no) 
    ability_to_absorb_loss_scrore  = get_absorb_loss_score(ability_to_absorb_loss)
    return  ability_to_absorb_loss_scrore - absorb_loss_constraint_target

def constraint3(weights):
    # Target Yield constraint (3.9% <= sum(weights * ytm) <= 4.8%)
    return  (max_target_yeild_target/100) - np.sum((weights) * (ytm/100)) 

def constraint5(weights):
    # Target Yield constraint (3.9% <= sum(weights * ytm) <= 4.8%)
    return  np.sum((weights) * (ytm)) - (min_target_yeild_target) 

def constraint6(weights):
    # Investment Horizon constraint (3.6 <= sum(weights * duration) <= 4.4)
    return   np.sum((weights) * duration) - min_investment_horizon_target 


# total sum of weight should be equal to 1 
def constraint4(weights):
    return  np.sum((weights)) - 1


    
# Initial guess for weights
initial_weights = [ (100/ num_rows)/100 for _ in range(num_rows)]
# print(initial_weights)
# print(np.sum(initial_weights))

# Constraints definition
constraints = (
                {'type': 'ineq', 'fun': constraint1},
                {'type': 'eq', 'fun': constraint2},
                {'type': 'ineq', 'fun': constraint3},
                {'type': 'eq', 'fun': constraint4},
                {'type': 'ineq', 'fun': constraint5},
                {'type': 'ineq', 'fun': constraint6},
    
              )

# Bounds for weights (between 0 and 2.5)
bounds = [(0,  2.5/100) for _ in range(num_rows)]
# print(bounds)

# Solve the optimization problem using GRG algorithm
result = minimize(objective_function, initial_weights, args=(duration, index_no, ytm, 1), method='SLSQP', bounds=bounds, constraints=constraints,  options = {"maxiter":1000})

# Extract the optimal weights
optimal_weights = result.x

print(result)

# Print the optimal weights
print("Optimal Weights:", optimal_weights)
print('SUM Optimal Weights: ', np.sum(optimal_weights))
```
